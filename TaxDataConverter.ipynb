```python
def get_price(entry):
    return float(entry.split('/')[0].split('$')[1])
    

def getMarketValue(produce):
    assert isinstance(produce,str), 'Please use a string value for produce'
    
    url  = 'https://www.amazon.com/s?k='+produce+'&i=wholefoods&ref=nb_sb_noss_2'

    page = requests.get(url, headers = HEADERS)
    soup = BeautifulSoup(page.content, features="lxml")
    
    print(soup)
    
    assert False

    souplist = [s.get_text() for s in soup.find_all('span', class_ = 'a-size-base a-color-secondary')]

    lblist = []
    for s in souplist:
        if 'lb' in s:
            lblist.append(get_price(s))
            
    
    ouncelist = []
    for s in souplist:
        if 'Ounce' in s:
            ouncelist.append(get_price(s)*16.)
            
    print(souplist)
    print(lblist)
    print(ouncelist)


    return min([min(lblist),min(ouncelist)])
    

    
```

```python
import os
from tqdm import tqdm
if not os.path.isdir('DeductionData'):
    os.mkdir('DeductionData')

for name in tqdm(names):
#     print(name)
    newdf = dfdata[dfdata['   Farm'] == name]
    newdict = {'Produce':[],'Quantity':[],'Market Value':[],'Deduction':[]}
#     print(np.char.strip(newdf['   Produce'].values.astype(str)))
    
    prodtypes = list(set(list(newdf['   Produce'].values.astype(str))))
    
#     print(prodtypes)
    for p in prodtypes:
        newnewdf = newdf[newdf['   Produce']==p]
        sumval = np.sum(newnewdf.iloc[:,13].values.astype(int))
        newdict['Produce'].append(p)
        newdict['Quantity'].append(sumval)
        mkval = getMarketValue(p)
        newdict['Market Value'].append(mkval*sumval)
        newdict['Deduction'].append(.5*mkval*sumval)
        
    dffarm = pd.DataFrame(newdict)
    dffarm.to_csv(os.path.join('DeductionData',name+' Deduction Data.csv'),index = False)
    
    
    
    #
    
    
newdf.iloc[:,[1,12,13,14]]

print(newdict)
```
